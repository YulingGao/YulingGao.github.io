```python
import pandas as pd
```

```python
data = pd.read_csv("master.csv")
```

```python
data
```

```python
data = data.drop(data.columns[[6,7,8,9,11]], axis=1)
data
```

```python
data['case_year'] = data.groupby(['year'])['suicides_no'].transform(sum)
data
```

```python
data['pop_year'] = data.groupby(['year'])['population'].transform(sum)
data
```

```python
data['case_sex']=data.groupby(['year','sex'])['suicides_no'].transform(sum)
data
```

```python
data['pop_sex']=data.groupby(['year','sex'])['population'].transform(sum)
data
```

```python
data['case_age_sex']=data.groupby(['year','sex','age'])['suicides_no'].transform(sum)
data
```

```python
data = data[data['year'] != 2016]
data
```

```python
data.to_csv('cleaned.csv')
```
